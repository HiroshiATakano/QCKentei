## 03-01 Mejora (mejora continua)

**Mejora** es la acción de hacer que las cosas sean mejores y, en la gestión de la calidad, consiste en mejorar la calidad, los costos, la productividad, etc., de los productos y servicios.

La **mejora continua** es la repetición de las actividades de mejora. Es importante que las actividades de mejora no se realicen una sola vez, sino que repitan y se mantengan de forma continua implementando el ciclo PDCA.


## 03-02 Historia QC

La historia QC: es un procedimiento para llevar a cabo actividades de mejora de manera efectiva y organizada, generalmente siguiendo ocho pasos.

### 1. Historia QC

Se difine un **problema** como la diferencia entre el estado actual y el estado ideal y el estado ideal, mientras que un **desafío** se define como la diferencia entre el estado actual y el estado futuro deseado.


```python
img = Image.open('./Data4/QCStory.png')
display(img)
```

### 2. Historia QC de resolución de problemas

Paso 1: Slección del tema

Paso 2: Compresión de la situación actual

* Basado en hechos y datos, no en intuición o experiencia.

Paso 3: Análisis de las causas

* Brainstorming / lluvia de ideas
* Diagrama de Ishikawa
* Factor impotante / Factor clave / Factor crítico
* Factor significativo / Factor relevante

Paso 4: Estudio y planificación de contramedidas

* Diagrama de árbol / Árbol de causas

Paso 5: Implementación y seguimiento de contramedidas

Paso 6: Confirmación de los efectos

* Gráfico
* Cuantitivo
* Volver a los pasos 3 a 5 y repetir.

Paso 7: Estandarización y consolidación de la gestión

Paso 8: Reflexión y acciones futuras

## 03-03 3Ms

Las 3Ms: Dentro de las actividades de mejora, se mencionan tres puntos de vista importantes para identificar problemas: la eliminación de Muri (sobrecarga), Mura (irregularidad) y Mura (desperdicio). Se les llama las 3Ms por sus iniciales.

* Muri (sobrecarga) se refeicere a algo que es difícil de hacer o llevar a cabo, algo que resulta trabajoso o imposible.

* Mura (irregularidad) se refiere a algo que es desigual, no uniforme o que tiene variaciones.

* Muda (desperdicio) se refiere a algo inútil o a un trabajo que no genera valor añadido.

### 3. Principio de Pareto / Ley de Pareto

Como se muestra en el diagrama, aunque haya muchos elementoss de clasificación, a menudo son los 2 o 3 principales los que tienen un gran impacto, y a esto se le llama el **Principio de Pareto**.


```python
from IPython.display import display
from PIL import Image
import gdown
import re

def get_file_id(url,out):
    # 正規表現でファイルIDを抽出
    match = re.search(r'/d/([a-zA-Z0-9_-]+)', url)

    if match:
        file_id = match.group(1)
        print("ファイルID:", file_id)
    else:
        print("ファイルIDが見つかりませんでした。")

    url = f"https://drive.google.com/uc?id={file_id}"

    output = out

    gdown.download(url, output, quiet=False)

    img = Image.open(out)
    display(img)
```

```python
url= 'https://drive.google.com/file/d/1GNzKFSsn8fruJWX3F3sL2BLnNLLJs0rF/view?usp=drive_link'
out = './Data4/QCStory.png'
get_file_id(url,out)
```
